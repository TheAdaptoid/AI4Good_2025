```python
import pandas as pd
import seaborn as sns
```

```python
# load datasets
rent_df = pd.read_csv("./data/JAX-Renter-Features.csv")
owner_df = pd.read_csv("./data/JAX-Owner-Features.csv")
print("Rent df shape:", rent_df.shape)
print("Owner df shape:", owner_df.shape)
rent_df.head(2)
```

```python
rent_df.corr()
```

```python
def normality_check(df: pd.DataFrame) -> pd.DataFrame:
    """Check normality of numerical columns in the dataframe using seaborn's histplot."""
    numeric_cols = df.select_dtypes(include=['number']).columns
    output = {'Column': [], 'Normality': []}
    for col in numeric_cols:
            normality = (df[col].mean() - df[col].median()) / df[col].std()
            output['Column'].append(col)
            output['Normality'].append(normality)
    return pd.DataFrame(output)

normality_results = normality_check(rent_df)
normality_results
```
